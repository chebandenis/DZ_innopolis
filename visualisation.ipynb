```python
# Отключаем ворнинги
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# Seaborn
import seaborn as sns
sns.set()

import plotly.express as px
```

Задача 1.

1. Сформировать массивданных с шагом 0.01 для функцииниже (с использованием пакета NumPy)и отобразить этиданные на экране в виде обычного графика в декартовых координатных осях с помощью библиотеки matplotlib

```python
def f(x):
    return np.cos(20*x)/(x+0.1)
```

```python
x = np.arange(0, 4, 0.01)

plt.figure(figsize=(10, 5))

plt.plot(x, f(x), linewidth=2, color='g', dashes=[8, 4], label='cos(20*x)/(x+0.1)')


plt.axis([0, 1.3 * np.pi, -5, 5])

plt.xticks(np.linspace(0, 1.3 * np.pi, 9), fontsize=12)  # Где сделать отметки
plt.yticks(fontsize=12)

plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$f(x)$', fontsize=20)

plt.title('cos(20*x)/(x+0.1)', fontsize=20)

plt.legend(fontsize=18, loc=0)

plt.show()
```

2. Отобразить тот же график с помощью библиотеки seaborn.

```python
sns.set(context='notebook', style='darkgrid', palette='dark', font='arial', font_scale=2, color_codes=False, rc=None)
sns.lineplot(x=x, y=f(x))
```

3. Отобразить тот же график с помощью библиотеки plotly

```python
px.line(x=x, y=f(x))
```

```python
df = pd.read_csv('car_data.csv')
df.head()
```

1. С использованием диаграммы boxplot визуализируйте зависимость числового признака от категориального.Ящики в диаграмме визуализировать горизонтально.

```python
plt.figure(figsize=(25, 10))


sns.boxplot(y='Body Style', x='Price ($)', data=df, orient='h')
plt.ylabel('Цена')
```

2. Постройте Scatterplot для двух числовыхп ризнаков из набора данных, отметив разными цветами категориальный признак.

```python
plt.figure(figsize=(25, 10))


sns.scatterplot(y='Annual Income', x='Price ($)', data=df, hue='Gender')
```

3. Постройте диаграмму Countplot для нескольких категориальных признаков (минимально взять 2 признака) из набора данных.Используйте библиотеку seaborn или plotly

```python
plt.figure(figsize=(25, 10))

sns.catplot(x='Transmission', col='Body Style', data=df, hue='Gender', kind='count')
```
