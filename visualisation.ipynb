```python
# Отключаем ворнинги
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# Seaborn
import seaborn as sns
sns.set()

import plotly.express as px
```

Задача 1.

1.

```python
def f(x):
    return np.cos(20*x)/(x+0.1)
```

```python
x = np.arange(0, 4, 0.01)

plt.figure(figsize=(10, 5))

plt.plot(x, f(x), linewidth=2, color='g', dashes=[8, 4], label='cos(20*x)/(x+0.1)')


plt.axis([0, 1.3 * np.pi, -5, 5])

plt.xticks(np.linspace(0, 1.3 * np.pi, 9), fontsize=12)  # Где сделать отметки
plt.yticks(fontsize=12)

plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$f(x)$', fontsize=20)

plt.title('cos(20*x)/(x+0.1)', fontsize=20)

plt.legend(fontsize=18, loc=0)

plt.show()
```

2.

```python
sns.set(context='notebook', style='darkgrid', palette='dark', font='arial', font_scale=2, color_codes=False, rc=None)
sns.lineplot(x=x, y=f(x))
```
